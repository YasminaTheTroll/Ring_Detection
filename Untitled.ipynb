```python
import math

img_bgr = cv2.imread("partial.jpg", cv2.IMREAD_COLOR)
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV_FULL)
img_hsv = cv2.resize(img_hsv, None, fx = 0.1, fy = 0.1)

low_orange = np.array([7,65,235])
high_orange = np.array([35,255,255])

mask = cv2.inRange(img_hsv, low_orange, high_orange)

def detect(img, low=np.array([7, 65, 235]), high=np.array([35, 255, 255])):
    img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV_FULL)
    

    mask = cv2.inRange(img_hsv, low, high)
    
    contours, hierarchy = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    out = cv2.cvtColor(mask, cv2.COLOR_GRAY2RGB)
    for c in contours:
        perimeter = cv2.arcLength(c, True)
        if perimeter == 0:
            continue
            
        area = cv2.contourArea(c)
    
        roundness = 4 * math.pi * area/(perimeter * perimeter)
        
        if roundness > 0.6:
            x,y,w,h = cv2.boundingRect(c)
            if w > 0.06 * img.shape[1]  and h > 0.03 * img.shape[0]:
                out = cv2.rectangle(out,(x,y),(x+w,y+h),(0,255,0),2)

    plt.imshow(out)
    print(x,y,w,h)
    print(area)

img_bgr = cv2.resize(img_bgr, None, fx = 0.1, fy = 0.1)
detect(img_bgr)
```
